```python
import pandas as pd
import matplotlib.pyplot as plt

dados = {
    'Mês': ['Janeiro', 'Fevereiro', 'Março', 'Abril'],
    'Consumo': [109, 150, 85, 93]
}

# Carregar dados
df = pd.DataFrame(dados)

# Visualizar as primeiras linhas dos dados para entender a estrutura
print(df.head())

# Plotar o gráfico
plt.figure(figsize=(10, 6))  # Define o tamanho da figura

# Aqui você pode personalizar o tipo de gráfico e os dados que deseja plotar
plt.plot(df['Mês'], df['Consumo'], marker='o', linestyle='--')

# Adicionar rótulos e título
plt.xlabel('Mês')
plt.ylabel('Consumo')
plt.title('Consumo ao longo dos Meses')

# Exibir o gráfico
plt.grid(True)  # Adiciona grade
plt.show()
```
